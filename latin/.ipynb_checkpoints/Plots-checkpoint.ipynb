# Plots for 2-Gram and 3-Grams based on various files
For all the plots, the same files were considered when considering n number of files. 
So, for the case of 2 files, the same two files were used for calculating the time for all the different combinations for plotting.

```r
library(ggplot2)
```

# Reading CSV containing times

```r
twoGrams <- read.csv("2Gram.csv")
threeGrams <- read.csv("3Gram.csv")
```

```r
names(twoGrams) <- c("Files","Time")
names(threeGrams) <- c("Files","Time")
```

# Plot for 2Grams for various files
For more than 300 files, our laptop gave runtime error saying the thread killed unexpectedly. So, we just used the first 300 files to plot this graph.

```r
ggplot(data = twoGrams, aes(y=Time, x = Files)) + geom_line() + geom_point() +
    ggtitle("2 Grams for Files using Spark")
```

# Plot for 3 Grams for Various Files
For 3 grams, the laptops maxed out at 40 files only since the amount of data generated in this case was huge.

```r
ggplot(data = threeGrams, aes(y=Time, x = Files)) + geom_line() + geom_point() +
    ggtitle("3 Grams for Files using Spark")
```

# Comparative Plot for 2Gram and 3Gram
For comparison sake, let's plot the run times for both the types of outputs together and see any relation between them.

```r
type1 <- c()
for(i in 1:length(twoGrams$Files)){
    type1 <- c(type1, "2Gram")
}
```

```r
type2 <- c()
for(i in 1:length(threeGrams$Files)){
    type2 <- c(type2, "3Gram")
}
```

```r
all_files <- c(twoGrams$Files, threeGrams$Files)
all_time <- c(twoGrams$Time, threeGrams$Time)
all_type <- c(type1, type2)
all_data <- data.frame(all_files,all_time,all_type)
names(all_data) <- c("Files","Time","Type")
```

```r
ggplot(data = all_data, aes(y = Time, x = Files, color=Type)) + geom_line() + geom_point() + 
    ggtitle("Comparison between 2grams and 3grams for different number of files")
```

# Inference
Based on the data collected for Spark running on a laptop, we can conclude the following things:
* The run time for 2grams is much less compared to 3grams as the number of files increases.
* The output size for the 2grams and 3grams is different by a large scale. For 2grams it is much less compared to the 3grams.
* Compared to Hadoop, the run time for large number of files is significantly lower which goes with the way Spark is supposed to work. It is expected to perform much better than Hadoop.
